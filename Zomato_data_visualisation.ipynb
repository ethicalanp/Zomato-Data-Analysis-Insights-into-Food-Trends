```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_data = pd.read_csv('/content/Zomato data .csv')
```

```python
zomato_data.info()
```

```python
zomato_data
```

```python
def handle_rate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

zomato_data['rate'] = zomato_data['rate'].apply(handle_rate)
print(zomato_data.head())
```

#Type of Restaurent

```python
zomato_data.head()
```

```python
sns.countplot(x=zomato_data['listed_in(type)'])
plt.xlabel("Type Of Resturent")
```

```python
grouped_data = zomato_data.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c='green',marker='o')
plt.xlabel("Type of Resturent ", c='red',size=20)
plt.ylabel('Votes',c='red',size=20)
```

```python
plt.hist(zomato_data['rate'],bins = 5)
plt.title("Ratings Distribustion")
plt.show()
```

#Mode Reciving Maximum ratings

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y ='rate', data = zomato_data)
```

```python
print(plt.colormaps())
```

```python
pivot_table = zomato_data.pivot_table(index = 'listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True , cmap='YlOrRd',fmt='d')
plt.title('Heatmap')
plt.xlabel("Online Order")
plt.ylabel('Listed In (Type)')
plt.show()
```
